### 1.2. Dati utilizzati

Abbiamo usato i dati provenienti dal dataset bibiografico che di seguito verrà linkato: https://raw.githubusercontent.com/dhdmch/2025-2026/refs/heads/main/data/lispod/data.csv.

Prima di tutto abbiamo strutturato il dataset mettendolo in una tabella su cui sono state riportate tutte le informazioni. I dati sono stati classificati nel seguente modo:
1) Id: link che rimanda al sito web o al database su cui si trovano gli articoli.
2) Titolo
3) Autori: chi ha scritto il documento.
4) Data di Pubblicazione: Anno o periodo in cui il documento è stato pubblicato.
5) Argomenti: di cosa parla il testo.
6) Basi dei Dati: si riferisce alle piattaforme, ai servizi di indicizzazione o ai repository accademici dove i documenti (articoli, pubblicazioni, ecc.) sono stati pubblicati, sono indicizzati o possono essere trovati. Ad esempio:  "Scopus", "DOAJ", o "Emerging Sources Citation Index", sono tutti nomi di database accademici.
7) DOI disponibili: Digital Object Identifier (DOI). Un DOI è un identificatore univoco e permanente assegnato a pubblicazioni digitali, come articoli di riviste, libri, capitoli di libri e altri tipi di documenti. La sua funzione principale è fornire un link persistente alla risorsa digitale, garantendo che essa sia sempre rintracciabile anche se la sua posizione (URL) cambia nel tempo. Se un DOI è disponibile per una pubblicazione, significa che è stato assegnato un identificatore standardizzato che ne facilita la citazione e il recupero.
8) Editori: l'azienda che si è occupata di investire nella pubblicazione e nella produzione di un certo numero di copie di un libro di un determinato autore.
9) Licenze della rivista: è l'insieme di termini e condizioni legali che regolano l'uso, la distribuzione e l'accesso ai contenuti pubblicati in una determinata rivista. Queste licenze stabiliscono se il contenuto è ad accesso aperto (open access), se può essere riutilizzato per scopi commerciali o non commerciali, e quali sono i diritti e le responsabilità degli utenti, degli autori e dell'editore. Ad esempio, una licenza "Creative Commons Attribution" (CC BY).
10) Rivista: se è una pubblicazione di una rivista.
11) Edizioni
12) Volume
13) Url Disponibile

Variabile: Questa è semplicemente il nome della colonna nel tuo dataset, come 'id', 'titolo', 'autori', 'data_pubblicazione', ecc.

Tipo: Indica il tipo di dato che la variabile contiene. Ad esempio, 'stringa' (testo), 'numerico' (numeri interi o decimali), 'data' (formato data), 'booleano' (vero/falso), ecc.

CREAZIONE TABELLA UNICA PER EDITORI, AUTORI E DATE DI PUBBLICAZIONI

```python
import pandas as pd

# Estrai i top 15 editori (già con nomi semplificati)
top_15_editori = df['editori'].value_counts().nlargest(15).rename('Numero di Pubblicazioni')

# Estrai i top 15 autori (già con nomi semplificati)
top_15_autori = df['autori'].value_counts().nlargest(15).rename('Numero di Pubblicazioni')

# Estrai i top 15 anni di pubblicazione
top_15_date = df['data_pubblicazione'].value_counts().nlargest(15).rename('Numero di Pubblicazioni')

# Converti le Series in DataFrame per poter aggiungere una colonna 'Categoria'
# Per una migliore presentazione, possiamo crearli come DataFrames indipendenti
# e poi unirli con concat o mostrarli separatamente se le lunghezze sono diverse.

# Se l'obiettivo è un'unica tabella che li elenca, li possiamo mettere in un dizionario e poi in un DataFrame
# Nota: le lunghezze potrebbero essere diverse, quindi il risultato potrebbe avere NaN dove non c'è corrispondenza.

# Per creare una singola tabella 'larga' che li contenga tutti e tre:
combined_df = pd.DataFrame({
    'Top 15 Editori': top_15_editori.index,
    'Editori Counts': top_15_editori.values,
    'Top 15 Autori': top_15_autori.index,
    'Autori Counts': top_15_autori.values,
    'Top 15 Anni Pubblicazione': top_15_date.index,
    'Anni Pubb. Counts': top_15_date.values
})

display(combined_df)
```
